```python
from bs4 import BeautifulSoup
from requests import get
from time import sleep
from random import randint
import pandas as pd
import itertools
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
headers = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'})
```

```python
precio = []
area = []
zona = []
descripcion = []
urls = []
```

```python
%%time

n_pages = 0

for page in range(0,50):
    n_pages += 1
    argenprop_url = 'https://www.argenprop.com/ph-alquiler-barrio-villa-urquiza-barrio-coghlan-barrio-saavedra-barrio-villa-ortuzar-hasta-25000-pesos'
    r = get(argenprop_url, headers=headers)
    page_html = BeautifulSoup(r.text, 'html.parser')
    house_containers = page_html.find_all('div', class_="listing__item")
    if house_containers != []:
        for container in house_containers:
            
            price = container.find_all('p')[0].text.strip()
            price_ = [int(price[s]) for s in range(0,len(price)) if price[s].isdigit()]
            price = ''
            for x in price_:
                price = price+str(x)
            precio.append(int(price))
            
            location = container.find_all('p', class_="card__location")[0].text.strip()
            zona.append(location)
            
            m2 = container.find_all('p', class_="card__common-data")[0].text.strip()
            area.append(m2)
            
            desc = container.find_all('p', class_="card__info")[0].text.strip()
            descripcion.append(desc)
            
            link = 'https://www.argenprop.com' + container.find_all('a')[0].get('href')
            urls.append(link)
            
    else:
        break
        
    sleep(randint(1,2))
    
print('You scraped {} pages containing {} properties.'.format(n_pages, len(urls)))
```

```python
cols = ['Zona', 'Precio', 'Caracteristicas', 'Descripcion', 'URL']

phs = pd.DataFrame({'Precio': precio,                    
                   'Caracteristicas': area,
                   'Zona': zona,
                    'Descripcion': descripcion,
                    'URL': urls,
                    })[cols]

phs.to_csv('phs_raw.csv')

phs = pd.read_csv('phs_raw.csv')
```

```python
phs = phs.drop_duplicates(['URL'])
```
